```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### Q2-1 : Let say sample data for the blood sugar of male patients are as 
“blood_sugar_male = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82,
129]”
 
Write a Python code to plot Histogram showing normal, pre-diabetic and 
diabetic patients distribution (Use 80-100: Normal; 100-125: Pre-diabetic and 80-
100: Diabetic). 
Also use Blood sugar samples of female patients as “blood_sugar_women = [67,
98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]”. 
Write a Python code to plot Histograms for both male and female patients

```python
# Creating data
blood_sugar_male = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82,129]
blood_sugar_male.sort()
blood_sugar_male
```

```python
# Creating dataset
df1 = pd.DataFrame(blood_sugar_male, columns=['blood_sugar'])

#Function to categorize the values
def find_range(df):
   if ( (df['blood_sugar']> 40) and (df['blood_sugar']< 100) ) :
     return 'Normal'
   elif ( (df['blood_sugar']> 100) and (df['blood_sugar']< 125) ):
     return 'Pre-Diabetic'
   elif (df['blood_sugar']> 125) :
     return 'Diabetic';

# Creating new Column
df1['Type'] = df1.apply(find_range, axis=1)

# Print the entire dataset
df1
```

```python
# Histogram plot
z=sns.histplot(data = df1, x = "blood_sugar", kde = True, bins = 20)
z.set_xlabel("Blood Sugar Range")
z.set_ylabel("No. Of Patients")
z.set_title('Blood Sugar Male')
```

```python
# Histogram plot with Legend
z=sns.histplot(data = df1, x = "blood_sugar", kde = False, hue = "Type",bins = 20, legend=True,alpha = 0.8, palette=["g", "C1", "r"])
z.set_xlabel("Blood Sugar Range")
z.set_ylabel("No. Of Patients")
z.set_title('Blood Sugar Male')
```

```python
# Histogram plot with 3 Range [0-100,100-125,125-200]
z=sns.histplot(data = df1, x = "blood_sugar", kde = False, hue = "Type",bins = [0,100,125,200], legend=True,alpha = 0.8, palette=["g", "C1", "r"])
z.set_xlabel("Blood Sugar Range")
z.set_ylabel("No. Of Patients")
z.set_title('Blood Sugar Male')
#z.set_xticks(range(3))
#z.set_xticklabels([0,100,125,200])
```
