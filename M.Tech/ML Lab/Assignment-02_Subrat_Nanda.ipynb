Q5 : Visualize the Bicycle Count in East, West during the month of March 2012

```python
#Converting DataType from String to DateTime
df["Date"]= pd.to_datetime(df["Date"])
df.head()
```

```python
# Filter by single month and Year #March 2012
df2 = df[df['Date'].dt.strftime('%Y-%m') == '2012-03']
df2

###Note: March 2012 has no data
```

```python
# Filter by single month and Year #March 2013
df2 = df[df['Date'].dt.strftime('%Y-%m') == '2013-03']
df2
```

```python
print(df2[df2.columns[1]].sum())
print(df2[df2.columns[2]].sum())
print(df2[df2.columns[3]].sum())
```

```python
x = list(df2.iloc[:,2:4].columns.values)
print(x);
```
