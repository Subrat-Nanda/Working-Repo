## Question-01

### Q1. Write a Python program to generate a data frame of 10 students with {Name, Gender, Age, CGPA, Marks, Mode of Exam, and Available for Weekend Class} as headers.

#### Explanation : 
Pandas DataFrame is a 2-dimensional labeled data structure like any table with rows and columns.

DataFrame() function is used to create a dataframe in Pandas.

#### Synatx : 
pandas.DataFrame(data, index, columns)

```python
# Python code demonstrate creating
# DataFrame from dict narray / lists
# By default addresses.

import pandas as pd

# initialize data of lists.
data = {'Name': ['Student 1', 'Student 2','Student 3','Student 4','Student 5','Student 6','Student 7','Student 8','Student 9','Student 10'],
		    'Gender': ['Male', 'Female','Male', 'Female','Male', 'Female','Male', 'Female','Male', 'Female'],
        'Age': [20, 21, 22, 20, 21, 22,20, 21, 22,20],
        'CGPA' : [4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5],
        'Marks' : [70,71,72,73,74,75,76,77,78,79],
        'Mode_of_Exam' : ['Online','Online','Online','Online','Online','Offline','Offline','Offline','Offline','Offline'],
        'Available_for_Weekend_Class' : ['Yes','No','Yes','No','Yes','No','Yes','No','Yes','No']
        }

# Create DataFrame
df = pd.DataFrame(data)

# Print the entire dataset
df
```

## Question-03

### Q3: Also insert one more column “Branch”. The entries in this column will be
“CSE” for students having CGPA greater than 7 and “Civil” for students having
CGPA in between 6 to 6.9

#### Explanation:
    Condition : CGPA between 6 - 6.9 Assign Branch as CVIL
                CGPA More than 7 Assign Branch as CSE

```python
# Function to assign the Branch based on Condition
def find_branch(df):
    if df['CGPA'] >= 7:
        return 'CSE'
    elif df['CGPA'] >= 6 and  df['CGPA'] <= 6.9 :
        return 'CVIL'
    else:
        return 'NA'

# Creating new Column
df['BRANCH'] = df.apply(find_branch, axis=1)

# Print the entire dataset
df
```

## Question-04

### Q4: Select Male students from the data frame with CGPA greater than 6.5.

#### Explanation : Filter data based on Gender and CGPA > 6.5

```python
# Filtering data Gender = Male and CGPA more than 6.5
result1_df = df[(df.CGPA > 6.5) & (df.Gender == 'Male')]

result1_df
```

## Question-05

####Q5 : List out female students interested for online mode of Exam and replace the corresponding entries with “Google Meet”

```python
#Replacking value of Mode_of_Exam from Online to Google Meet
result2_df = df[(df.Gender == 'Female') & (df.Mode_of_Exam == 'Online')].replace(to_replace="Online",value="Google Meet")

result2_df
```
