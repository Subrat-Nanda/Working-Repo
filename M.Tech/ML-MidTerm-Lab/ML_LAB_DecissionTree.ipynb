```python
import pandas as pd 
df=pd.read_csv('/content/Titanic_DataSet/titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch',	'Ticket',	'Cabin',	'Embarked'],axis='columns',inplace= True)
```

```python
df.head()
```

```python
target = df['Survived']
target
```

It is used to convert non numerical label to numerical label

```python
from sklearn.preprocessing import LabelEncoder
```

```python
L_Sex = LabelEncoder()
```

```python
df['Sex_New'] = L_Sex.fit_transform(df['Sex'])
df
```

```python
df_new = df.drop(['Sex','Survived'],axis='columns')
df_new
```

```python
df_new.isnull()
```

```python
df_new.isnull().sum()
```

```python
df_new.Age = df_new.Age.fillna(df_new.Age.mean())
df_new.head()
```

```python
df_new.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_new, target, test_size=0.2)
```

```python
df_new.shape
```

```python
print('X_train  : ' + str( len(X_train) ) )
print('X_test  : ' + str( len(X_test) ) )
print('y_train  : ' + str( len(y_train)) )
print('y_test  : ' + str( len(y_test)) )
```

```python
from sklearn import tree
model =tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pre = model.predict(X_test)
```

```python
y_pre
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
matirx =  confusion_matrix(y_test, y_pre, labels=[1,0])
```

```python
print('Confusion Matrix : \n' , matirx)
```

```python
matirx_C =  classification_report(y_test, y_pre, labels=[1,0])
```

```python
print('Classification Report : \n' , matirx_C)
```
